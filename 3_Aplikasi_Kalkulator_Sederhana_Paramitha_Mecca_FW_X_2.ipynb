```python
panjang = int(input("Masukan panjang dari persegi panjang (cm): "))
lebar =int(input("Masukkan lebar dari persegi (cm):"))

luas = panjang * lebar

print(f"\nLuas dari persegi panjang yaitu {luas} cm^2")
```

```python
r = int(input("Masukan jari jari dari kerucut:"))
t = int(input("Masukan tinggi dari kerucut:"))

volume = 1/3 * 22/7 * r**2 * t

print(f"\nvolume dari kerucut yaitu {volume} cm^3")
```
